```python
import csv
import pandas as pd
import numpy as np
```

```python
import folium
from folium import plugins
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# import data
data_2018_30days = pd.read_csv('Data_2018_2103_2004_with_added_columns.csv', nrows= 1000)
```

```python
data_2018_30days.head()
```

```python
# find minimum of longitudes
lon_min = data_2018_30days['Longitude'].min()
lon_min
```

```python
# find maximum of longitudes
lon_max = data_2018_30days['Longitude'].max()
lon_max
```

```python
# find average of longitudes
lon_avg = (lon_min+lon_max)/2
lon_avg
```

```python
# find minimum of latitudes
lat_min = data_2018_30days['Latitude'].min()
lat_min
```

```python
# find maximum of latitudes
lat_max = data_2018_30days['Latitude'].max()
lat_max
```

```python
# find average of latitudes
lat_avg = (lat_min+lat_max)/2
lat_avg
```

```python
# set the map location
m = folium.Map([lat_avg, lon_avg], zoom_start=14)
m
```

```python
# mark each parking location as a point
for index, row in data_2018_30days.iterrows():
    folium.CircleMarker([row['Latitude'], row['Longitude']],
                        radius=3,
                        popup=row['BlockfaceName'],
                        fill_color="#000000", 
                       ).add_to(m)
```

```python
# convert to (n, 2) nd-array format for heatmap
parkingArr = data_2018_30days[['Latitude', 'Longitude']].as_matrix()

# plot heatmap
m.add_child(plugins.HeatMap(parkingArr, radius=20))
m
```
